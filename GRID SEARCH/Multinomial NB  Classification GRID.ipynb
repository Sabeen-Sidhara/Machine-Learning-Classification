```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,dtype=int, drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.drop("age",axis=1)
```

```python
dataset
```

```python
dataset["classification_yes"].value_counts()
```

```python
dataset.columns
```

```python
independent=dataset[['bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv', 'wc',
       'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent.shape
```

```python
independent
```

```python
dependent=dataset[[ 'classification_yes',]]
```

```python
from sklearn.model_selection import train_test_split    
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import GridSearchCV                
param_grid = {'alpha': [1,2,3,4,5,5],
              'force_alpha':[True,False],
              'fit_prior':[True,False],        
              'class_prior':[None]}

grid = GridSearchCV(MultinomialNB(), param_grid, refit = True, verbose = 3,cv=5,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_

grid_prediction=grid.predict(x_test)

```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_prediction)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_prediction)
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_prediction,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
